```python
def get_neighbors(gene_list,graph,node_dict):
    """
    """
    nbr_lists = []
    for gene in gene_list:
        nbr_lists.append(sorted([gene]+[node.name for node in graph.neighbors(node_dict[gene])]))
        
    return nbr_lists
```
