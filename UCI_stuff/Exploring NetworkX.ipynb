```python
import networkx as nx
#import matplotlib.pylab as pylab
import sys,os
import StringIO
```

```python
def make_graph(rounds=1):
    graph = nx.Graph()
    r = 0
    while rounds-r > 0:
        # complete
        graph.add_edge('A%s.1' % (r), 'B%s.1' % (r))
        graph.add_edge('B%s.1' % (r), 'C%s.1' % (r))
        graph.add_edge('C%s.1' % (r), 'A%s.1' % (r))
        #complete
        graph.add_edge('A%s.2' % (r), 'B%s.2' % (r))
        graph.add_edge('B%s.2' % (r), 'C%s.2' % (r))
        graph.add_edge('C%s.2' % (r), 'A%s.2' % (r))
        #complete
        graph.add_edge('A%s.3' % (r), 'B%s.3' % (r))
        graph.add_edge('B%s.3' % (r), 'C%s.3' % (r))
        graph.add_edge('C%s.3' % (r), 'A%s.3' % (r))
        # incomplete
        graph.add_edge('A%s.4' % (r), 'B%s.4' % (r))
        graph.add_edge('B%s.4' % (r), 'C%s.4' % (r))
        r+=1
    return graph
```

```python
g = make_graph(5)
g['B1.4']
```

## Test a few subgraph extractors for their ability to scale

```python
g1 = make_graph(1)
g10 = make_graph(10)
g100 = make_graph(100)
g1000 = make_graph(1000)
```

```python
def edges_in_K_n(n):
    edges = (n*(n - 1.0))/2.0
    return edges

def get_completes_nx(graph):
    subgraphs = nx.connected_component_subgraphs(graph)
    completes = [x for x in subgraphs if len(x.edges()) == edges_in_K_n(len(x.nodes()))]
    return completes
```

```python
%timeit -r 10 g1_cmplts = get_completes_nx(g1)
```

```python
%timeit -r 10 g1_cmplts = get_completes_nx(g10)
```

```python
%timeit -r 10 g1_cmplts = get_completes_nx(g100)
```

```python
%timeit -r 10 g1_cmplts = get_completes_nx(g1000)
```
