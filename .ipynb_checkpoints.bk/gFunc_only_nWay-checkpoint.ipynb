## 2.1: Develope ability to test for "$n$-way" sets 

### Equation to determine correct number of edges ($E_n$) in a completely connected graph with $n$ nodes ($K_{n}$)

$E_{n} = n(n-1)/2$

\[[source](http://en.wikipedia.org/wiki/Complete_graph#Properties)\]   

A completely connected subgraph represents a $n$-by-$n$ ($n$ 1-to-1 ortholog relationships) which is what I am seeking.

### Function to find $E_n$

```python
def K_n_edges(n):
    edges = (n*(n - 1.0))/2.0
    return edges
```

### Function to collect subgraphs with complete connectedness

```python
def get_complete_subgraphs(graph):
    subgraphs = nx.connected_component_subgraphs(graph)
    completes = [x for x in subgraphs if len(x.edges()) == K_n_edges(len(x.nodes()))]
    return completes
```

## 2.2: Assess functions on faked up example data

### Function to generate representative graphs

`make_graph()` makes the number and pattern of 3-node graphs defined in the `while` loop in multiples of the value given to `rounds`.

```python
def make_graph(rounds=1):
    graph = nx.Graph()
    r = 0
    while rounds-r > 0:
        # complete
        graph.add_edge('A%s.1' % (r), 'B%s.1' % (r))
        graph.add_edge('B%s.1' % (r), 'C%s.1' % (r))
        graph.add_edge('C%s.1' % (r), 'A%s.1' % (r))
        #complete
        graph.add_edge('A%s.2' % (r), 'B%s.2' % (r))
        graph.add_edge('B%s.2' % (r), 'C%s.2' % (r))
        graph.add_edge('C%s.2' % (r), 'A%s.2' % (r))
        #complete
        graph.add_edge('A%s.3' % (r), 'B%s.3' % (r))
        graph.add_edge('B%s.3' % (r), 'C%s.3' % (r))
        graph.add_edge('C%s.3' % (r), 'A%s.3' % (r))
        # incomplete
        graph.add_edge('A%s.4' % (r), 'B%s.4' % (r))
        graph.add_edge('B%s.4' % (r), 'C%s.4' % (r))
        r+=1
    return graph
```

```python
g1 = make_graph(1)
nx.draw_graphviz(g1)
title('make_graph(1)')
```

```python
g3 = make_graph(3)
nx.draw_graphviz(g3)
title('make_graph(3)')
```

```python
g1 = make_graph(1)
g10 = make_graph(10)
g100 = make_graph(100)
g1000 = make_graph(1000)
```

### Correctness of `get_complete_subgraphs()`:

We will test `get_complete_subgraphs()` correctness using the `g3` graph set created and drawn earlier to make sure that the not fully connected subgraphs are removed.

```python
g3_complete = get_complete_subgraphs(g3)

# nx.draw_graphviz() requires a single graph to draw not a list of graphs
# recombine subgraphs to allow drawing
g3_complete = nx.union_all(g3_complete)
```

```python
nx.draw_graphviz(g3_complete)
title('get_complete_subgraphs(g3)')
```
