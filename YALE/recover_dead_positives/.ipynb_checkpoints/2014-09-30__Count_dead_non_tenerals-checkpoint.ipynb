# Getting an estimate of the number of DEADs we have to deal with 

```python
from spartan.utils import spreadsheets as ss
```

```python
ug_spring_summer = "/home/gus/Dropbox/uganda data/2014_Collection_Sheets_Spring-Summer/2014_full_surveyreport_20140820/Dissection data.xls"
```

```python
ug_wb = ss.get_workbook(ug_spring_summer)
```

```python
ug_ws = ss.get_worksheets(ug_wb)
```

```python
ws = ug_ws[0]
```

```python
ws.col_values(7)[:7]
```

```python
def get_rownum_if(col, func, target):
    rownums = []
    for i, row in enumerate(col):
        if func(row, target):
            rownums.append(i)
    return rownums
```

```python
def f(rowval,target):
    if unicode(target).upper() in unicode(rowval).upper():
        return True
    else:
        return False
```

```python
get_rownum_if(ws.col_values(7),f, 'dead')[:5]
```

```python
def get_col_hit_sets(worksheet, func, target):
    hit_sets = []
    for i in range(worksheet.ncols):
        hit_sets.append(set(get_rownum_if(worksheet.col_values(i), func, target)))
        
    return hit_sets
        
```

```python
get_col_hit_sets(worksheet=ws, func=f, target='dead')[:8]
```

```python
len(frozenset().union(*get_col_hit_sets(worksheet=ws, func=f, target='dead')))
```

```python
def estimate_hits_per_sheet(worksheet, func, target):
    return len(frozenset().union(*get_col_hit_sets(worksheet=worksheet, func=func, target=target)))
```

```python
def get_estimates_by_sheet(workbook, func, target):
    s = []
    for ws in workbook.sheets():
        est = estimate_hits_per_sheet(worksheet=ws, func=func, target=target)
        s.append(est)
        print "{0}:\t{1}".format(ws.name, est)
    
    print "TOTAL:\t{0}".format(sum(s))
```

# Final Tally

```python
get_estimates_by_sheet(ug_wb, f, 'dead')
```
