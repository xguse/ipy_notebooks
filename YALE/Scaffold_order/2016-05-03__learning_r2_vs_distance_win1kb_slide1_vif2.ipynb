# Imports

```python
import seaborn as sns
sns.set_context('talk')
sns.set_style('whitegrid')

import pandas as pd
import numpy as np

import scipy.stats as sstats
import statsmodels as sm

import sklearn as skl
```

# Functions

```python
def load_and_recode(path):
    ld = pd.read_csv(path, sep="\t").rename(columns={"R^2":"R2"})
    ld["DISTANCE"] = abs(ld.POS2 -ld.POS1)
    
    return ld
```

# Basic Exploration

```python
mean_ld_by_dist = ld.groupby("DISTANCE").mean().reset_index()
```
