# Implementation:

## Imports:

```python
# imports
import pandas as pd
```

## File paths:

```python
# define paths to files
base_dir = "/home/gus/remote_mounts/louise/data/genomes/glossina_fuscipes/annotations/functional/GfusI1.1_pre/argot2_out/"


argot_in = base_dir + "argot_functional_annotations_ts150.tsv"
argot_h5 = base_dir + "argot_functional_annotations_ts150.h5"
```

# Read data into `pandas`:

```python
argot_df = pd.read_table(argot_in, engine='python', sep='\t')
```

```python
argot_df.head()
```

#  Test write using assignment to manually created `HDFStore` object:

```python
# create our storage object
argot_store = pd.io.pytables.HDFStore(argot_h5)
```

```python
# register the dataframe with the store
argot_store['dataframe'] = argot_df
```
