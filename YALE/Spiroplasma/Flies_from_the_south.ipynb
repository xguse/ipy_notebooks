```python
def fuzzy_filter(item, targets):
    """Returns boolean if filter is successfull on ``item``.

    Args:
        item (str): string in dataframe column to be filtered
        targets (list): list of location codes we want from the database.

    Returns:
        boolean
    """

    hit = process.extractOne(item, targets)
    return hit[-1] >= 80

def mask_by_fuzzy(targets, db_col):
    """Creates extraction mask to pull out targets from db based on fuzzy matching.
    
    Args:
        targets (list): list of location codes we want from the database.
        db_col (Series): dataframe column representing the vaules to be filtered.
    

    Returns:
        pandas mask
    """
    
    mask = db_col.apply(fuzzy_filter, targets=targets)
    
    return mask
    
```
