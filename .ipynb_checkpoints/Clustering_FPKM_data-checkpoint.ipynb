# Load some data

```python
Ag_file = '/home/gus/Dropbox/common/projects/Aa_Ag_Cq_As/RNAseq_results/bio/dunnw/data/4mosqs_mg_RNAseq/gtf_strict/gene_fpkm_dumps/Ag.fpkm'
AgFPKMs = pd.read_table(Ag_file, sep=' ', header=0, index_col=0)
AgGTF = '/home/gus/Dropbox/common/projects/Aa_Ag_Cq_As/RNAseq_results/bio/dunnw/data/4mosqs_mg_RNAseq/gtf_strict/cuffmerge_Ag_00_0.Ag_04_0.Ag_06_0.Ag_08_0.Ag_10_0/merged.gtf'

# Allows translating XLOCs to Ensembl Names
Ag_xloc2vb = cufflinks.get_XLOC_to_nearest_ref_map(AgGTF)

# Load list of orthologs that I am interested in
Ag_orthos = '/home/gus/Dropbox/common/projects/Aa_Ag_Cq_As/gfunc_stuff/prelim_gene_analysis/ecr_OR_insect_20130903/ptci_1_0/ecr_OR_insect_meanPTCI_thresh_1_gene_names.AGAP.txt'
Ag_orthos = [x.rstrip('\n') for x in open(Ag_orthos,'rU')]
```

```python
def mplot(df,clsIDs,log=True,legnd=False,ttl=False,save=False):
    id_list = sorted(set(clsIDs))
    id_num = len(id_list)
    
    color_set = [cm.spectral(float(x)/id_num)[:-1] for x in id_list]

    
    dimens = int(np.ceil(np.sqrt(len(id_list))))
    fig,axes = plt.subplots(nrows=dimens, ncols=dimens)
    ax = cartesian([range(dimens),range(dimens)])
    
    if ttl:
        fig.suptitle(ttl,fontsize=18,y=1.02)
    
    for i in id_list:
        members = len([x for x in clsIDs==i if x])
        frac = round(float(members)/len(df),2)
        j,k = ax[i][0],ax[i][1]
        group_color = [color_set[i]] * members
        #group_color = color_set[i]
        
        df.ix[clsIDs==i].T.plot(ax=axes[j,k],
                                title='ID: %s | members: %s | frac: %s'%(i,members,frac),
                                legend=0,logy=log,lw=2,
                                color=group_color)
        
        dfs = pd.DataFrame(df.ix[clsIDs==i].median())
        dfs.plot(ax=axes[j,k],color='grey', ls='-', lw=10, alpha=0.9,label='median',legend=0)
        
    tight_layout(pad=0, w_pad=0, h_pad=0)
    #autoscale()
    
    # delete axes that have no lines plotted
    for i,x in enumerate(axes):
        for j,y in enumerate(x):
            if len(getp(y,'lines')) < 1:
                axes[i][j].set_visible(False)
    
    if save:
        out_path = save
        print 'saving figure as: %s' % (out_path)
        savefig(out_path)
```

# Prepare different treatments of the ortholog data

```python
# Full FPKM matrix dump from cummerbund
X = AgFPKMs

# pull out the orthologs I am looking for after translating their names into XLOCs
xloc_orthos = [Ag_xloc2vb[x] for x in Ag_orthos]

# Nothing done to it, pure FPKMs from cummerbund
Xo = X.ix[xloc_orthos]

Xo1 = Xo+1

# Log10 transformed (plus one to prevent logging of zeros and subsequent universe to implosion)
XoLg = np.log10(Xo+1) 

# z-type standardized (gene-wise; aka row-wise)
XoSTD = spd.stddf(Xo,1)
```

# Cluster raw FPKMs with pearson coeff

```python
nclusters = 25
npass = 10000
```

```python
Xo_cluster_ids,Xo_error,Xo_solution_freq = kcluster(Xo,nclusters=nclusters,npass=npass, method='a', dist='c')
```

```python
Xo_error
```

```python
Xo_solution_freq
```

```python
counts,bins,patches = hist(Xo_cluster_ids,bins=range(nclusters+1),normed=1,histtype='bar',color='c',alpha=0.7)
xlabel('cluster id')
ylabel('proportion of genes per cluster')
locs,labs = xticks(bins[:-1]+0.5, range(nclusters))
autoscale()
save_figure('Xo_cluster_ids.hist.png')
```

```python
mplot(Xo1,Xo_cluster_ids,log=1,ttl='Xo; Pearson; plotted as Xo+1,log-y')
```

```python
mplot(XoSTD,Xo_cluster_ids,log=0,ttl='Xo; Pearson; plotted as XoSTD')
```
