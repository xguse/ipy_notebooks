# Functions

```python
def get_nodes_by_attr_values(g, attr_name, attr_values):
    """Return generator of nodes whose values for `attr_name` are in `attr_values`."""
    attr = attr_name
    vals = set(attr_values)
    
    return (n for n,d in g.nodes_iter(data=True) if d[attr] in vals)
    
```
