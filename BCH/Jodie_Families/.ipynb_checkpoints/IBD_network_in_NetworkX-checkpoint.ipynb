# Functions

## Table manipulation

```python
def top_cands_prune_columns(df):
    """Return pruned table with one row per top candidate."""
    col_map = {"family_name": "family",
               "WES/WGS": "seq_type",
               "Family_size": "fam_size",
               "Likely inheritance model": "inherit_mode",
               "short phenotype description": "phenotype",
               "Merckanalysis top candidates": "top_candidates",
              }
    
    keep = ["family","seq_type","fam_size","inherit_mode","top_candidates","phenotype"]
    
    # prune columns and clean some row values
    pruned_table = df.rename(columns=col_map)
    pruned_table = pruned_table[keep].copy().dropna()
    pruned_table["top_candidates"] = pruned_table["top_candidates"].apply(lambda i: i.upper().strip().replace(' ','').split(','))
    pruned_table["family"] = pruned_table["family"].apply(lambda i: i.upper().replace("FAMILY","").strip())
    
    return explode_col(df=pruned_table, idx_col="family", exp_col="top_candidates")[keep]

def explode_col(df, idx_col, exp_col):
    """Return dataframe where `exp_col` is exploded and then merged back into `df` using `idx_col`.
    
    `exp_col` must contain existing lists.
    `idx_col` should be suitible for use as a primary key (unique).
    """
    keep = [col for col in df.columns.values if col != exp_col]
    
    # explode the top_candidates row lists into new dataframe
    exploded = df[[idx_col,exp_col]].set_index(idx_col)
    exploded = exploded.to_dict()[exp_col]
    
    idx_value_pairs = []
    for idx, values in exploded.items():
        idx_value_pairs.extend([[idx, value] for value in values])

    
    exploded = pd.DataFrame(idx_value_pairs, columns=[idx_col,exp_col]).set_index(idx_col)
    
    # join the exploded df with the rest of the data we want to keep
    remerged = pd.merge(left=df[keep].set_index(idx_col), right=exploded,
                      how="inner", 
                      on=None, left_on=None, right_on=None, 
                      left_index=True, right_index=True,)
    
    return remerged.reset_index(idx_col).copy()
```

## Working with the graph

```python
def get_nodes_by_attr_values(g, attr_name, attr_values):
    """Return generator of nodes whose values for `attr_name` are in `attr_values`."""
    attr = attr_name
    vals = set(attr_values)
    
    return (n for n,d in g.nodes_iter(data=True) if d[attr] in vals)

def check_nodes_for_paths(g, nodes):
    """Return dictionary keyed with directional node tuples and valued with the shortest path if any between purmuations of the `nodes` list."""
    all_shortest = nx.all_pairs_shortest_path(g)
    node_pairs = itertools.permutations(nodes, 2)
    
    has_path = {}
    for source, target in node_pairs:
        try:
            path = all_shortest[source][target]
            has_path[(source,target)] = path
        except KeyError:
            has_path[(source,target)] = None
            
    return has_path

def get_interactions_from_path(g, path):
    """Return interleaved list of nodes and interactions of the path provided."""
    sources = path[:-1]
    targets = path[1:]
    
    interactions = []
    for s, t in zip(sources, targets):
        
        if g.node[s]['name'] not in interactions:
            interactions.append(g.node[s]['name'])
        
        interactions.append(g[s][t]['interaction'])
        interactions.append(g.node[t]['name'])
    
    return interactions
```

### Can we find paths between any of the genes? 

```python
has_path = check_nodes_for_paths(g=ibd, nodes=top_genes_ibd)
for pair, path in has_path.items():
    if path is not None:
        print("{pair}: {path}".format(pair=pair, path=[(ibd.node[n]["name"],n) for n in path]))
```
